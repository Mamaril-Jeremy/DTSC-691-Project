## Load and Put Together Datasets

```python
# Load the datasets
youtube_videos_usa = pd.read_csv("../data/USvideos.csv", encoding='utf-8').sample(n=16000)
youtube_videos_canada = pd.read_csv("../data/CAvideos.csv", encoding='utf-8').sample(n=16000)
youtube_videos_mexico = pd.read_csv("../data/MXvideos.csv", encoding='latin-1').sample(n=16000)


# Merge the two datasets
youtube_videos = pd.concat([youtube_videos_usa, youtube_videos_canada, youtube_videos_mexico])

# View the first 5 rows
youtube_videos.head()
```

## Data Cleaning and Preprocessing

```python
def clean_data(df):
    # Clean columns
    df.columns = df.columns.str.strip().str.lower()
    
    # Check for NULLs and Drop Any
    print("Check for NULLs and Drop Them")
    print(df.isnull().sum())
    df = df.dropna().reset_index(drop=True)  
    print("\n\nNo more NULLs left")
    print(df.isnull().sum())
    print("\n\n")
    
    # Clean textual columns
    textual_cols = ['title', 'channel_title', 'tags', 'description']
    for col in textual_cols:
        df[col] = df[col].astype(str).str.replace('\n', '').str.strip()
    
    # Clean tags
    df['tags'] = df['tags'].apply(lambda x: x.split('|') if x != '[none]' else [])
    
    # Clean dates
    df['trending_date'] = pd.to_datetime(df['trending_date'], format='%y.%d.%m')
    df['publish_time'] = pd.to_datetime(df['publish_time'])
    
    # Make binary columns binary
    bool_cols = ['comments_disabled', 'ratings_disabled', 'video_error_or_removed']
    for col in bool_cols:
        df[col] = df[col].astype(bool)
    
    return df 

youtube_videos = clean_data(youtube_videos)  
youtube_videos.head()
```

```python
# Filter out rows

# Undefined engagement
youtube_videos = youtube_videos[youtube_videos['views'] > 0]

# Cannot define engagement rate with no ratings
youtube_videos = youtube_videos[~(youtube_videos['ratings_disabled'])]

youtube_videos.head()
```

```python
# Calculate raw engagement rate 
youtube_videos["engagement_rate_raw"] = (
    youtube_videos["likes"] + 
    youtube_videos["comment_count"]
) / youtube_videos["views"]
```

```python
# Extract Post Hour
youtube_videos['publish_hour'] = youtube_videos.loc[:, 'publish_time'].dt.hour 

# Extract Day of the Week
youtube_videos['publish_day'] = youtube_videos['publish_time'].dt.dayofweek
youtube_videos['publish_day_name'] = youtube_videos['publish_time'].dt.day_name()
```

```python
# Determine Tag Count
youtube_videos['tag_count'] = youtube_videos['tags'].apply(lambda x: len(x))
```

```python
# Title Sentiment Modeling
import torch
from transformers import pipeline

# Define device and dtype (safe defaults that work on CPU or GPU)
device = 0 if torch.cuda.is_available() else -1
torch_dtype = torch.float16 if torch.cuda.is_available() else torch.float32

title_pipe = pipeline(
    'sentiment-analysis',
    model='cardiffnlp/twitter-roberta-base-sentiment-latest',
    device=device,
    torch_dtype=torch_dtype,
)

def convert_to_polarity(res):
    label, score = res['label'], res['score']
    if label == 'positive': return  score
    if label == 'negative': return -score
    return 0.0

titles = youtube_videos['title'].fillna('').astype(str).tolist()
results = title_pipe(titles, batch_size=64, truncation=True)
youtube_videos['title_sentiment'] = [convert_to_polarity(r) for r in results]
```

## Model Preparation

```python
features = [
    'category_id',
    'comments_disabled',
    'publish_hour',
    'publish_day',
    'tag_count',
    'title_sentiment',
    'description_sentiment'
]
```

```python
X = youtube_videos[features]
y = youtube_videos["engagement_rate_raw"]

X_train, X_test, y_train, y_test = train_test_split(X, y, train_size=0.8)
```

```python
# Selected the following numeric columns
numeric_columns = ['title_sentiment', 'description_sentiment', 'tag_count']

# Selected the following categorical columns
categorical_columns = ['category_id', 'comments_disabled', 'publish_hour']

# Selected the following ordinal columns
ordinal_columns = ['publish_day']
```

```python
# Create pipeline
from sklearn.pipeline import make_pipeline
from sklearn.impute import SimpleImputer
from sklearn.preprocessing import StandardScaler, OneHotEncoder, OrdinalEncoder
from sklearn.compose import ColumnTransformer

# Creates numeric pipeline 
# Scales the data to reduce dimensionality
numeric_pipeline = make_pipeline(
    StandardScaler()
)

# Creates categorical pipeline 
categorical_pipeline = make_pipeline(
    OneHotEncoder(drop='first', sparse_output=False)
)

# Creates ordinal pipeline 
ordinal_pipeline = make_pipeline(
    SimpleImputer(strategy='most_frequent').set_output(transform='pandas'),
    OrdinalEncoder(categories='auto')
)

# Uses ColumnTransformers to pass the data through the different pipelines to handle the variety of data types
column_transformer = ColumnTransformer(
    [('num', numeric_pipeline, numeric_columns),
     ('cat', categorical_pipeline, categorical_columns),
     ('ord', ordinal_pipeline, ordinal_columns)]
)

# Fit and transform the data through the ColumnTransformers
X_train_transformed = column_transformer.fit_transform(X_train, y_train)
X_test_transformed = column_transformer.transform(X_test)
```

## Model Creation

```python
# Create RandomForest Regressor
random_forest_model = RandomForestRegressor()
random_forest_model.fit(X_train_transformed, y_train)

y_pred = random_forest_model.predict(X_test_transformed)

print("R² Score:  ", r2_score(y_test, y_pred))
print("MAE:       ", mean_absolute_error(y_test, y_pred))
print("RMSE:      ", root_mean_squared_error(y_test, y_pred))
cv_scores_random_forest = cross_val_score(random_forest_model, X_train_transformed, y_train, scoring='neg_mean_squared_error', cv=3)
print(cv_scores_random_forest)
```

```python
# Create XGBoost Regressor Model
xgb_model = XGBRegressor()
xgb_model.fit(X_train_transformed, y_train)

y_pred = xgb_model.predict(X_test_transformed)

print("R² Score:  ", r2_score(y_test, y_pred))
print("MAE:       ", mean_absolute_error(y_test, y_pred))
print("RMSE:      ", root_mean_squared_error(y_test, y_pred))
```

```python
# Create Linear Regression Model
linear_model = LinearRegression()
linear_model.fit(X_train_transformed, y_train)

y_pred = linear_model.predict(X_test_transformed)

print("R² Score:  ", r2_score(y_test, y_pred))
print("MAE:       ", mean_absolute_error(y_test, y_pred))
print("RMSE:      ", root_mean_squared_error(y_test, y_pred))
```
